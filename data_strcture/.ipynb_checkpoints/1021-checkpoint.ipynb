```python
class Queue:
    def __init__(self, lens):
        self.len = lens
        self.queue = list([-1] * lens)
        self.head = 0
        self.rear = 0
        self.amount = 0
    def isEmpty(self):
        return self.head == self.rear
    def isFull(self):
        return (self.head ==self.len-1 and self.rear == 0)
    def canPush(self):
        return self.head < self.len 
    def canPop(self):
        return self.rear < self.len  and self.rear < self.head
    
    def push(self, t):
        if self.canPush():
            self.amount += 1
            self.queue[self.head] = t
            self.head +=1
            return True
        else:
            print("Queue is Gone")
            return False
            
    def pop(self):
        if self.canPop():
            self.amount -= 1
            out = self.queue[self.rear]
            self.rear += 1
            return out
        else:
            print("Can't pop")
            return False
    def printQueue(self):
        temp = self.rear
        out = list([0] * self.amount)
        for i in range(self.amount):
            out[i] = self.queue[temp]
            temp += 1
        return out[::-1]
    def back(self):
        return self.queue[self.rear]
    def front(self):
        return self.new

a = Queue(3)

a.printQueue()

for i in [1,3, 5]:
    a.push(i)

print(a.isFull())

for i in [1,3, 5]:
    print(a.pop())

a.push(2)

print(a.isFull())
```

```python

class Node:
    def __init__(self, data, nexts = None):
        self.data = data
        self.next = nexts

class LinkedList:
    def __init__(self, head = None):
        self.head = head
        
    def isEmpty(self):
        return self.head == None
    
    def insertAtFirstNode(self, newNode):
        if self.isEmpty():
            self.head = newNode
        else:
            newNode.next = self.head
            self.head = newNode
        
    def removeFirstNode(self):
        if self.isEmpty():
            print("is Empty")
            return false
        else:
            out = self.head
            self.head = self.head.next
            return out
    
    def insertNode(self, curNode, newNode):
        temp = curNode.next
        curNode.next = newNode
        newNode.next = temp
    
    def insertAtLastNode(self, newNode):
        if self.isEmpty():
            self.head = newNode
        else:
            yes = self.head
            while(yes.next != None):
                yes = yes.next
            yes.next = newNode
            
    def removeLastNode(self):
        if self.isEmpty(): 
            print("Empty, can't remove")
            return False
        else:
            yes = self.head
            while(yes.next != None):
                yes = yes.next
            out = yes.next
            yes.next = None
            return out
        
    def __len__(self):
        if self.isEmpty(): return 0
        yes = self.head
        out = 1
        while(yes.next != None):
            yes = yes.next
            out += 1
        return out
    def printList(self, resend = False):
        if self.isEmpty(): return "List([])"
        yes = self.head
        out = 'List(['
        while(yes != None):
            out += (str(yes.data) + ', ') 
            yes = yes.next
        out = out[0:-2] + "])"
        if resend: return out
        print(out)
        
    def concatenate(self, other):
        self.insertAtLastNode(other.head)
    def inverse(self):
        if self.isEmpty(): return LinkedList()
        yes = self.head
        prev = None
        while(yes != None):
            temp = yes.next
            data = yes.data
            prev = Node(data, nexts = prev)
            yes = temp
        return LinkedList(prev)
    

a = LinkedList()

for i in range(7):
    a.insertAtLastNode(Node(i))

 

c = a.inverse()

a.printList()
c.printList()
```
