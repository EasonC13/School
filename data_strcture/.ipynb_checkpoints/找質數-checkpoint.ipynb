```python
import time
start = time.time()
prime_nums = []
dur_time = []
i = 1
while(True):
    i += 1
    trigger = 0
    for j in prime_nums:
        if i%j==0:
            trigger = 1
    if trigger == 0:
        prime_nums.append(i)
        temp_time = time.time()
        dur_time.append(temp_time - start)
        print("%fs, \t %d"%(dur_time[-1], i))
        start = temp_time
```
