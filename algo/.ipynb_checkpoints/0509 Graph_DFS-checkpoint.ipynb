```python
class Node:
    def __init__(self, name, value = None, edge = None, color = "White", parent = None):
        self.name = name
        self.value = value
        self.edge = edge if edge else []
        self.color = color
        self.parent = None
    def addEdge(self, newEdge):
        self.edge.append(newEdge)
    def setColor(self, color):
        self.color = color
    

nodes = ["u,v,w,z,x,y"]
nodes = nodes[0].replace(" ","").split(",")

nodeDict = {}
for node in nodes:
    nodeDict[node] = Node(node)

def getNodesEdge(nodes):
    out = []
    for node in nodes.split(","):
        out.append(nodeDict[node])
    return out


nodeDict['u'].edge += getNodesEdge("v,x")
nodeDict['v'].edge += getNodesEdge("y")
nodeDict['w'].edge += getNodesEdge("y,z")
nodeDict['z'].edge += getNodesEdge("z")
nodeDict['x'].edge += getNodesEdge("v")
nodeDict['y'].edge += getNodesEdge("x")

class Graph:
    def __init__(self, nodeDict):
        self.nodeDict = nodeDict
        
G = Graph(nodeDict)

def DFS(G):
    nodeDict = G.nodeDict
    G.times = 0
    G.optEdge = {}
    for key in nodeDict.keys():
        if nodeDict[key].color == "White":
            nodeDict[key].parent = None
            G.times += 1
            nodeDict[key].d = G.times
            DFS_Visit(nodeDict[key],G)
            G.times += 1
            nodeDict[key].f = G.times

def DFS_Visit(node, G):
    node.color = "Gray"
    for subnode in node.edge:
        if subnode.color == "White":
            G.optEdge["%s-%s"%(node.name, subnode.name)] = "T"
            G.times +=1
            subnode.parent = node
            subnode.d = G.times
            DFS_Visit(subnode, G)
            G.times += 1
            subnode.f = G.times
        elif subnode.color == "Gray":
            G.optEdge["%s-%s"%(node.name, subnode.name)] = "B"
        elif subnode.d > node.d:
            G.optEdge["%s-%s"%(node.name, subnode.name)] = "F"
        else:
            G.optEdge["%s-%s"%(node.name, subnode.name)] = "C"
    node.color = "Black"

DFS(G)

from graphviz import Digraph
dot = Digraph(comment='The Round Table')

for key in G.nodeDict.keys():
    node = G.nodeDict[key]
    dot.node(G.nodeDict[key].name, G.nodeDict[key].name + "%d/%d"%(node.d, node.f))
    for edge in G.nodeDict[key].edge:
        value = (G.optEdge["%s-%s"%(G.nodeDict[key].name, edge.name)])
        dot.edge(G.nodeDict[key].name, edge.name, value)

dot
```
